```python
''''QC. [Marks 5] Ignoring the dose attribute, compute Euclidean distance between the first and
second observation (row) using the following formula:
√( ∑ i=1n (x i −y i ) 2 ), where x and y represent two observations (rows). x i and y i represent the values of
ith variable of observation x and y, respectively. n is the total no. of variables.
Note: Before computing the Euclidean distance, normalize the vectors correspond to observation 1
and 2 using min-max normalization method (described below).
- Using min-max normalization, normalized value (nv 1,1 ) for observation 1 and variable 1 can be
computed as follows:
nv 1,1 = (v 1,1 -min(v ,1 ))/(max(v ,1 )-min(v ,1 )), where v 1,1 is the value given or observation 1 and variable
1; v ,1 reprsents all the values for variable 1 across all the observations (rows).'''

def norm_minmax(a):
    return (a-a.min())/(a.max()-a.min())

def euclidean_dist(a,b):
    return np.sqrt(sum((a-b)**2))

r1_frame = liver_frame.iloc[0,:-1]
r2_frame = liver_frame.iloc[1,:-1]
norm_r1= norm_minmax(r1_frame)
norm_r2 = norm_minmax(r2_frame)
euc_dist = euclidean_dist(norm_r1,norm_r2)
print('euclidean distance between r1 and r2 :', euc_dist)
```

```python
'''QA. [Marks 2] Compute variance of each row (probe-id) and remove the probe-id(s) for which the
variance is <=0.2 in order to create a new dataframe object where all the probe-ids will have
variance > 0.2. If you don’t find any probe-id having variance <=0.2, display the message “no such
probe-id is found!!!”, else print the resultant datadframe object.'''

dna_frame = pd.read_csv('data_mRNA_subset.txt', sep='\t')
high_variance_dna = dna_frame.drop(dna_frame.index[dna_frame.var(axis=1) <=0.2], axis=0)
print(high_variance_dna.head())
```

```python

'''QB. [Marks 5] Use the original dataframe object to compute Spearman Rank Correlation coefficient
between the first two probe-ids using the following formula:
Step I: For each probe-id, find the ranks for each individual column. Suppose, r 1 and r 2 is the rank
vectors for probe-id 1 and probe-id 2, respectively.
Step II: Compute d vector which is the difference between two rank vectors (r 1 and r 2 ).
Step III: Now, sum all of d-squared values.
Step IV: Use the aforementioned formula to compute Spearman Rank Correlation coefficient. n is
the total no. of columns that we have in the given dataset (i.e. 63).'''
n=dna_frame.shape[1]-1 # (Removing ID_REF column)
df_rank = dna_frame.rank(axis=1)
diff_rank = df_rank.iloc[0,:]-df_rank.iloc[1,:]
sq_diff_val = sum(diff_rank**2)
sp_rank = 1-(6*sq_diff_val/(n*(n**2-1)))
print('spearman rank correleation : ',sp_rank)
```
